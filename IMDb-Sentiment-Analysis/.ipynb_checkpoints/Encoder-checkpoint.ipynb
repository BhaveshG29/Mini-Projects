```python
import pandas as pd
import numpy as np
```

### Loading and Standarizing Text

```python
data_csv = (pd.read_csv("data.csv", header=None, sep=".").drop(columns=[1]).values)

data_list = list(text.lower().replace(",", "").replace("!", "").replace(".", "") for text in data_csv[:,0])

data_list, type(data_list[0])
```

### Tokenizing

```python
doc_tokens_list = list(doc.split() for doc in data_list)
doc_tokens_list
```

### Creating Personalized Vocabulary with mapping

```python
vocabulary_set = []
for token in doc_tokens_list:
    for word in token:
        vocabulary_set.append(word)
vocabulary_set = set(vocabulary_set)

vocabulary = {"UNK":0}
for i, word in enumerate(vocabulary_set):
    vocabulary.update({word: i+1})
vocabulary
```

### Creating the One-Hot Encoded Bag of Words Matrix

```python
N = len(doc_tokens_list) # Number of Documents
V = len(vocabulary) # Size of Vocabulary OR Number of Distinct Words

BoW_matrix = np.zeros((N,V)) #(10, 32)

for i, doc_tokens in enumerate(doc_tokens_list):
    for word in doc_tokens:
        if(word in vocabulary):
            col_idx = vocabulary[word]
            BoW_matrix[i, col_idx] += 1

print(f"Sparsity = {100*(1 - BoW_matrix.mean()):.2f}% | BoW-Matrix Shape: {BoW_matrix.shape}")
BoW_matrix
```

```python
TF_matrix = BoW_matrix/(BoW_matrix.sum(axis=1).reshape(-1,1)) # For this Dataset

df = (BoW_matrix>0).sum(axis=0).reshape(-1,1) #Number of Documents containing w_i  word

IDF_vector = 1 + np.log((N+1)/(df+1)) #Using Smooth IDF

TF_matrix, IDF_vector
```

```python
TF_IDF_matrix = (TF_matrix * IDF_vector.T)
v_d_norm = (np.sqrt(((TF_IDF_matrix**2).sum(axis=1)))).reshape(-1,1)
TF_IDF_matrix = TF_IDF_matrix/v_d_norm #L2 Normalization

TF_IDF_matrix
```
