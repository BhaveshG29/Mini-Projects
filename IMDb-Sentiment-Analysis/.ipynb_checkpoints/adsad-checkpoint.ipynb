```python
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
import pandas as pd
import time
```

```python
dataset = pd.read_csv("IMDB Dataset.csv")

dataset.head(10)
```

```python
review_df = dataset.iloc[:,0].values
labels = dataset.iloc[:,1].replace({"negative":0, "positive":1}).values.astype(np.float64)

doc_review = list(user_review.replace("<br /><br />", "") for user_review in review_df) #Documents

doc_review[0:2], labels[0:2]
```

```python
count_vectorizer = CountVectorizer(lowercase=True, ngram_range=(1,2)) #Bi-gram due to ngram_range from uni to bi
 
BoW_matrix = count_vectorizer.fit_transform(doc_review) #BoW_Matrix

#Vocabulary Vector of Bi-Grams
feature_names = count_vectorizer.get_feature_names_out()
```

```python
BoW_matrix.shape, feature_names.shape
```

```python
#Calculating frequency of each word
word_count = np.array(BoW_matrix.sum(axis=0)).flatten()

#Getting indices that would sort the counts (ascending)
sorted_idx = np.argsort(word_count)

#Defining your thresholds
k = 100  #Removes top-k most frequent words, often they are stop-words like "the", "and", "a"
p = 500  #Removes bottom 500 rarest words 

#Slicing the indices to keep the middle ground
# We remove 'p' from the start and 'k' from the end
keep_idx = sorted_idx[p : len(sorted_idx) - k]

#Filtering the Matrix and the Feature Names
BoW_matrix = BoW_matrix[:, keep_idx]
feature_names = feature_names[keep_idx]

#Calculating Sparisity
density = BoW_matrix.nnz / (BoW_matrix.shape[0] * BoW_matrix.shape[1])
sparsity = 100 * (1 - density)

print(f"Shrunk Vocab: {len(feature_names)}")
print(f"Shrunk BoW Matrix Size: {BoW_matrix.shape}")
print(f"Sparsity of BoW Matrix: {sparsity:.2f}%")
```

```python
N = BoW_matrix.shape[0]

shuffled_idx = np.random.permutation(N)
train_size = int(0.8 * N)

train_idx, test_idx = shuffled_idx[:train_size], shuffled_idx[train_size:]

X_train, X_test = BoW_matrix[train_idx], BoW_matrix[test_idx]

y_train, y_test = labels[train_idx].reshape(-1,1), labels[test_idx].reshape(-1,1)

X_train[:5, :5].toarray(), y_train[0:5], X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
def sigmoid(z):
    return (1/(1+np.exp(-z)))

def BCE(g, a):
    result = (a*np.log(g) + (1-a)*np.log(1-g)).mean()
    return -result

def accuracy(g,a):
    g_preds = (g>=0.5).astype(int)
    result = 100 * (g_preds == a).mean()
    return result
```

```python
alpha = 0.5
epochs = 250 #100

N = X_train.shape[0]
m = X_train.shape[1]
np.random.seed(1324)
W = np.random.randn(m,1) * 0.01
b = np.zeros((1,1))

X_csc = X_train.tocsc()

loss_hist = []
acc_hist = []

start_time = time.time()
for epoch in range(epochs):
    logits = X_csc@W + b
    y_preds = sigmoid(logits)

    error = y_preds - y_train
    dW = (1/N) * X_csc.T @ error
    db = (error).mean()

    W = W - alpha*dW
    b = b - alpha*db

    loss = BCE(y_preds, y_train)
    acc = accuracy(y_preds, y_train)

    loss_hist.append(loss)
    acc_hist.append(acc/100)
    
    if(epoch%50 == 0):
        print(f"Epoch: {epoch:3d} | Loss: {loss:.4f} | Accuracy: {acc:.2f}%")
        
end_time = time.time() - start_time

print(f"\nTime Taken to Train: {end_time:.3f} seconds")
print(f"Final Loss: {loss_hist[-1]:.4f}")
print(f"Final Accuracy: {acc_hist[-1]*100:.2f}%")
```

```python
X_csc = X_test.tocsc()

logits = X_csc@W + b
y_preds = sigmoid(logits)

loss = BCE(y_preds, y_test)
acc = accuracy(y_preds, y_test)

print(f"Test Loss: {loss:.4f} | Test Accuracy: {acc:.2f}%")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,8))
ax1 = plt.subplot(1,2,1)
ax2 = plt.subplot(1,2,2)

ax1.plot(np.arange(len(loss_hist)), loss_hist, c="r", label="Loss Curve")
ax1.scatter(len(loss_hist), loss, c="orange", marker="*", s=100, zorder=5, edgecolor="k")
ax1.legend()

ax2.plot(np.arange(len(acc_hist)), acc_hist, c="g", label="Accuracy Curve")
ax2.scatter(len(acc_hist), acc/100, c="b", marker="*", s=100, zorder=5, edgecolor="k")
ax2.legend()

plt.show()
```
